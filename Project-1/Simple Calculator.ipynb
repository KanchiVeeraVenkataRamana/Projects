```python
from tkinter import*
import math
import tkinter.messagebox

root=Tk()
root.geometry('650x400+300+300')
root.title('Scientific Calculator')
switch=None

#buttons number functions
def btn_1():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'1')
    
def btn_2():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'2')
    
def btn_3():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'3')
    
def btn_4():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'4')
    
def btn_5():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'5')
    
def btn_6():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'6')
    
def btn_7():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'7')
    
def btn_8():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'8')
    
def btn_9():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'9')
    
def btn_0():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,'0')
    

#buttons operation functions
def btn_p():
    pos=len(disp.get())
    disp.insert(pos,'+')
    
def btn_s():
    pos=len(disp.get())
    disp.insert(pos,'-')
    
def btn_ml():
    pos=len(disp.get())
    disp.insert(pos,'*')
    
def btn_di():
    pos=len(disp.get())
    disp.insert(pos,'/')
    
    
# buttons trignometry fuctions
def asin():
    try:
        ans=float(disp.get())
        if switch is True:
            ans=math.asin(math.radians(ans))
        else:
            ans=math.asin(ans)
            disp.delete(0,END)
            disp.insert(0,str(ans))
    except:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
            
def acos():
    try:
        ans=float(disp.get())
        if switch is True:
            ans=math.acos(math.radians(ans))
        else:
            ans=math.acos(ans)
            disp.delete(0,END)
            disp.insert(0,str(ans))
    except:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
            
def atan():
    try:
        ans=float(disp.get())
        if switch is True:
            ans=math.atan(math.radians(ans))
        else:
            ans=math.atan(ans)
            disp.delete(0,END)
            disp.insert(0,str(ans))
    except:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
        

#buttons on power,module,and more functions
def pow():
    pos=len(disp.get())
    disp.insert(pos,'**')
    
def dot():
    pos=len(disp.get())
    disp.insert(pos,'.')
    
def br():
    pos=len(disp.get())
    disp.insert(pos,'(')
    
def bl():
    pos=len(disp.get())
    disp.insert(pos,')')
    
def cl():
    pos=len(disp.get())
    disp.insert(pos,'')
   
def mo():
    pos=len(disp.get())
    disp.insert(pos,'%')

def eq():
    try:
        ans=disp.get()
        ans=eval(ans)
        disp.delete(0,END)
        disp.insert(0,ans)
    except Exception:
        tkinter.messagebox.showerror('Value error! check your operator and operand')

def log():
    try:
        ans=float(disp.get())
        ans=math.log10(ans)
        disp.delete(0,END)
        disp.insert(0,str(ans))
    except Exception:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
        
def sqrt():
    try:
        ans=float(disp.get())
        ans=math.sqrt(ans)
        disp.delete(0,END)
        disp.insert(0,str(ans))
    except Exception:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
        
        
def round():
    try:
        ans=float(disp.get())
        ans=round(ans)
        disp.delete(0,END)
        disp.insert(0,str(ans))
    except Exception:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
        
def ln():
    try:
        ans=float(disp.get())
        ans=math.log(ans)
        disp.delete(0,END)
        disp.insert(0,str(ans))
    except Exception:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
        
        
def fact():
    try:
        ans=fact(disp.get())
        ans=math.factorial(ans)
        disp.delete(0,END)
        disp.insert(0,str(ans))
    except Exception:
        tkinter.messagebox.showerror('Value error! check your operator and operand')
        
def btn_pi():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,str(math.pi))
    
def btn_ex():
    if disp.get()=='0':
        disp.delete(0,END)
    pos=len(disp.get())
    disp.insert(pos,str(math.e))
    
def back():
    pos=len(disp.get())
    display=str(disp.get())
    if display=='':
        disp.insert(0,'0')
    elif display=='':
        disp.insert(0,'0')
    elif display=='0':
        pass
    else:
        disp.delete(0,END)
        disp.insert(0,display[0:pos-1])
        
def conv():
    global switch
    if switch is None:
        switch=True
        conv['text']='Deg'
    else:
        conv['text']='Rad'

disp=Entry(root,font='Verdana 20',fg='Black',bg='mistyrose',bd=4,justify=RIGHT)
disp.insert(0,'0')
disp.pack(expand=True,fill=BOTH)


#buttons row-1
r1=Frame(root,bg='#000000')
r1.pack(expand=TRUE,fill=BOTH)
pibtn=Button(r1,text='π',font='segoe 18',relief=GROOVE,bd=0,command=btn_pi,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
fabtn=Button(r1,text='x!',font='segoe 18',relief=GROOVE,bd=0,command=fact,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
sinbtn=Button(r1,text='sin',font='segoe 18',relief=GROOVE,bd=0,command=asin,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
cosbtn=Button(r1,text='cos',font='segoe 18',relief=GROOVE,bd=0,command=acos,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
tanbtn=Button(r1,text='tan',font='segoe 18',relief=GROOVE,bd=0,command=atan,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn1=Button(r1,text='1',font='segoe 23',relief=GROOVE,bd=0,command=btn_1,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn2=Button(r1,text='2',font='segoe 23',relief=GROOVE,bd=0,command=btn_2,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn3=Button(r1,text='3',font='segoe 23',relief=GROOVE,bd=0,command=btn_3,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btnp=Button(r1,text='+',font='segoe 23',relief=GROOVE,bd=0,command=btn_p,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)


#botton row-2
r2=Frame(root,bg='#000000')
r2.pack(expand=TRUE,fill=BOTH)
e_btn=Button(r2,text='e',font='segoe 18',relief=GROOVE,bd=0,command=btn_ex,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
sqrt_btn=Button(r2,text='√x',font='segoe 18',relief=GROOVE,bd=0,command=sqrt,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
sinh_btn=Button(r2,text='sin-1',font='segoe 18',relief=GROOVE,bd=0,command=asin,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
cosh_btn=Button(r2,text='cos-1',font='segoe 18',relief=GROOVE,bd=0,command=acos,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
tanh_btn=Button(r2,text='tan-1',font='segoe 18',relief=GROOVE,bd=0,command=atan,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn4=Button(r2,text='4',font='segoe 23',relief=GROOVE,bd=0,command=btn_4,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn5=Button(r2,text='5',font='segoe 23',relief=GROOVE,bd=0,command=btn_5,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn6=Button(r2,text='6',font='segoe 23',relief=GROOVE,bd=0,command=btn_6,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btns=Button(r2,text='-',font='segoe 23',relief=GROOVE,bd=0,command=btn_s,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)


#bottons row-3
r3=Frame(root,bg='#000000')
r3.pack(expand=True,fill=BOTH)
conv_btn=Button(r3,text='Rad',font='segoe 18',relief=GROOVE,bd=0,command=conv,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
round_btn=Button(r3,text='round',font='segoe 18',relief=GROOVE,bd=0,command=round,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
ln_btn=Button(r3,text='ln',font='segoe 18',relief=GROOVE,bd=0,command=ln,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
log_btn=Button(r3,text='log',font='segoe 18',relief=GROOVE,bd=0,command=log,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
pow_btn=Button(r3,text='x^y',font='segoe 18',relief=GROOVE,command=pow,bd=0,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn7=Button(r3,text='7',font='segoe 23',relief=GROOVE,bd=0,command=btn_7,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn8=Button(r3,text='8',font='segoe 23',relief=GROOVE,bd=0,command=btn_8,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn9=Button(r3,text='9',font='segoe 23',relief=GROOVE,bd=0,command=btn_9,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btnml=Button(r3,text='*',font='segoe 23',relief=GROOVE,bd=0,command=btn_ml,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)


#bottons row-4
r4=Frame(root,bg='#000000')
r4.pack(expand=True,fill=BOTH)
mod_btn=Button(r4,text='%',font='segoe 18',relief=GROOVE,bd=0,command=mo,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
bl_btn=Button(r4,text=' ( ',font='segoe 18',relief=GROOVE,bd=0,command=bl,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
br_btn=Button(r4,text=' ) ',font='segoe 18',relief=GROOVE,bd=0,command=br,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
dot_btn=Button(r4,text='.',font='segoe 18',relief=GROOVE,bd=0,command=dot,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn_clear=Button(r4,text='C',font='segoe 18',relief=GROOVE,bd=0,command=cl,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
del_btn=Button(r4,text='del',font='segoe 23',relief=GROOVE,bd=0,command=back,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btn0=Button(r4,text='0',font='segoe 23',relief=GROOVE,bd=0,command=btn_0,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btneq=Button(r4,text='=',font='segoe 23',relief=GROOVE,bd=0,command=eq,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)
btnd=Button(r4,text='/',font='segoe 23',relief=GROOVE,bd=0,command=btn_di,fg='white',bg='#333333').pack(side=LEFT,expand=True,fill=BOTH)

root.mainloop()
```
